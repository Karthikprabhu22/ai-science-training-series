```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
tot_time = []
for i in [1,2,4,8,16]:
    train_acc, train_loss, val_acc, val_loss, time_per_ep = np.loadtxt("metrics_{}.dat".format(i), unpack=True)
    tot_time.append(sum(time_per_ep))
    plt.figure(i)
    plt.plot(val_acc, label = "Validation accuracy")
    plt.plot(train_acc, label = "Training accuracy")
    plt.xlabel("#Epoch")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.savefig("Accuracy_{}.pdf".format(i))
```

```python
tot_time
```

```python
plt.plot([1,2,4,8,16],tot_time)
plt.xlabel("#Nodes")
plt.ylabel("Total time")
plt.savefig("time.pdf")
```
