```python
training_report = "10110015.output"
```

```python
import pandas as pd
import re
with open(training_report) as f:
    lines = f.readlines()
li = []
for line in lines:
    if(re.match(r'^Finished step ', line)):
        x = re.search(r'^Finished step (?P<step>\d*) of (?P<tot_steps>\d*) in epoch (?P<epoch>\d*),loss=(?P<loss>[\d.]*), acc=(?P<acc>[\d.]*) \((?P<ips>[\d.]*)', line)
        li.append(x.groupdict())
df = pd.DataFrame(li)
```

```python
import numpy as np
x = np.array(df["step"].values[:-100],dtype=np.float32)
y = np.array(df["loss"].values[:-100],dtype=np.float32)
z = np.array(df["acc"].values[:-100],dtype=np.float32)
```

```python
import matplotlib.pyplot as plt
plt.plot(y, label="Loss")
plt.plot(z, label="Accuracy")
plt.legend()
plt.xlabel("Step")
```
